```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url="https://www.foxterciaimobiliaria.com.br/construtora/162518/melnick-even-desenvolvimento-imobilirio"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")

lista = []

for link in soup.findAll("a", {'class': 'ajaxLoad'}):
    x = link.get("href")
    if "/empreendimento/" in x:
        lista.append("https://www.foxterciaimobiliaria.com.br" + x)
```

```python
for i in range(len(lista)):
    print([i, (i+1)/len(lista)])
           
    url = lista[i]

    html_content = requests.get(url).text

    soup = BeautifulSoup(html_content, "lxml")

    table = soup.find("table", {"class": "table-detail"}).findAll("tr")

    seg = table[0].text[10:-1]
    bai = table[1].text[8:-1]
    cid = table[2].text[8:-1]
    emp = soup.find("div", {"class": "product-details"}).findAll("h2")[0].text.replace("\n",'').replace('\t','')
    
    codigo = [int(re.findall("\d+", item.text)[0]) for item in soup.findAll('span', {'class': 'value codigo'})]
    tipo = [item.text for item in soup.findAll('span', {'class': 'value tipo'})]
    area = [float(item.text[:-2]) for item in soup.findAll('span', {'class': 'value areaPrivativa'})]
    preco = [int(item.text[3:].replace('.','')) for item in soup.findAll('span', {'class': 'value preco'})]
    link = [('https://www.foxterciaimobiliaria.com.br/imovel/' + str(item)) for item in codigo]
    empreendimento = [emp] * len(codigo)
    bairro = [bai] * len(codigo)
    cidade = [cid] * len(codigo)
    
    data = {'codigo':  codigo,
        'empreendimento': empreendimento,
        'area': area,
        'preco': preco,
        'tipo': tipo,
        'bairro': bairro,
        'cidade': cidade,
        'link': link
        }

    df = pd.DataFrame (data, columns = ['codigo','empreendimento','area','preco','tipo','bairro','cidade','link'])

    if i == 0:
        df_final = df
    else:
        df_final = pd.concat([df_final, df])
            
```
